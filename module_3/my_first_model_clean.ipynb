# Загрузка Pandas и очистка данных

```python
# Загружаем необходимые библиотеки
import pandas as pd 
import numpy as np
import random
```

```python
df = pd.read_csv('main_task.csv') # создаем датафрейм
```

```python
df.info()
```

```python
df.drop(['ID_TA', 'URL_TA','Reviews'], axis = 1, inplace=True) # удаляем ненужные столбцы
```

```python
# убирем лишние символы из Cuisine Style
df['Cuisine Style'] = df['Cuisine Style'].str.replace('[' , '')
df['Cuisine Style'] = df['Cuisine Style'].str.replace(']' , '')
df['Cuisine Style'] = df['Cuisine Style'].str.replace("'" , '')
df['Cuisine Style'] = df['Cuisine Style'].str.replace(" " , '')
```

```python
#Напишем функцию,которая заместит пропуски в показателе кухни по принципу 5 наиболее популярных кухонь
def fill_csn(row):
    if pd.isnull(row['Cuisine Style']):
        return str(df['Cuisine Style'].dropna().explode().value_counts().head(5).index)[8:-19]
    return row['Cuisine Style']
```

```python
df['Cuisine Style']=df.apply(fill_csn,axis=1)# замещаем пропуски
```

```python
df['Cuisine Style'] = df['Cuisine Style'].str.replace('[' , '')# убирем лишние символы из Cuisine Style, переводим в list
df['Cuisine Style'] = df['Cuisine Style'].str.replace(']' , '')
df['Cuisine Style'] = df['Cuisine Style'].str.replace("'" , '')
df['Cuisine Style'] = df['Cuisine Style'].str.replace(" " , '')
df['Cuisine Style'] = df['Cuisine Style'].apply(lambda x: x.split(','))
```

```python
# создаем функцию для замены переменных Price Range на числовой параметр
def fill_PR(row):
    if pd.isnull(row['Price Range']):
        return np.random.randint(1,4)  # создаем случайный генератор для NaN от 1 до 3
    if row['Price Range'] == '$':
        return 1
    if row['Price Range'] == '$$ - $$$':
        return 2
    if row['Price Range'] == '$$$$':
        return 3
```

```python
df['Price Range']=df.apply(fill_PR,axis=1)
```

```python
df['Number of Reviews'].fillna(df['Number of Reviews'].median(), inplace=True)# заполняем пропуски в Number of Reviews
```

```python
df.info()
```

```python
df_CS=(df['Cuisine Style'].str.get_dummies(sep=','))# создаем dummy переменные
df_CS
```

```python
df_city=pd.get_dummies(df['City'],drop_first=True)# создаем dummy переменные
```

```python
df_city
```

```python
df_df_City=pd.merge(df, df_city, left_index=True, right_index=True)#объединяем датафреймы
```

```python
df_CS_City=pd.merge(df_df_City, df_CS, left_index=True, right_index=True)#объединяем датафреймы
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df_df_City.drop(['Restaurant_id', 'Rating','City','Cuisine Style'], axis = 1)
y = df['Rating']
```

```python
X.info()
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
